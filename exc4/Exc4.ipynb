```python
#11
data=pd.DataFrame({"Cname":['wipro','Cogniziant','TCS','Capgemini','virtusa'],'profit':[-120000,200000,-500000,-400000,300000]})
df=pd.DataFrame(data)
print(df)
print("-----UPDATED DATAFRAME----")
df['profit']=df['profit'].apply(lambda x:x>0)
print(df)
```

```python
data1={'eid':[1,2,3,4],'ename':["Bob",'Sam','Theethu','Vishnu'],'Stipend':[10000,5000,15000,20000]}
print("-----FIRST DATAFRAME----")
df1=pd.DataFrame(data1)
print(df1)
print("-----SECOND DATAFRAME----")
data2={'eid':[1,2,3,4],'designation':['Engineer','Senior employee','Intern employee','Engineer']}
df2=pd.DataFrame(data2)
print(df2)
dataframe=pd.merge(df1,df2,how='inner',on='eid')
print("-----MERGED DATAFRAME----")
print(dataframe)
```
