```python
import matplotlib.pyplot as plt
import numpy as nup
menMeans=nup.array([22, 30, 35, 35, 26])
womenMeans=nup.array([25, 32, 30, 35, 29])
width = 0.35
ind = nup.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.ylabel('Scores')
plt.xlabel('person')
plt.title('Scores by person')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.bar(ind + 0.00, menMeans, color = 'b', width = 0.25)
ax.bar(ind+ 0.25, womenMeans, color = 'g', width = 0.25)
plt.legend(labels=['Men', 'Women'])
plt.show()
```
